```python
orden_aic_bic=[]

# Iteramos sobre el coeficiente AR entre  0-3
for p in range(5):
  # Iteramos sobre el coeficioente MA entre 0-3
    for q in range(5):
      	# creamos y entrenamos el modelo ARMA(p,q) cuando sea posible 
        for d in range(5):
           for s in range(5):
            for t in range(8,13):
                try:
                    model = SARIMAX(c3_tmed['1981':]['prom'].dropna(),order=(p,q,d),seasonal_order=(s,0,0,t),trend='c')

                    results = model.fit()
        
                           # Adjuntamos a la lista los IC
                    orden_aic_bic.append((p,q,d,s,t,results.aic, results.bic,results.hqic,results.llf,results.mae,results.mse,results.sse))
                except:
                          ## Sino es posible, retornamos un NAN
                    orden_aic_bic.append((p,q,d,s,t,None,None,None,None,None,None,None))

#c4_tmed.info()
                
# DataFrame de orden_aic_bic
orden_df = pd.DataFrame(orden_aic_bic, 
                        columns=['p','q','d','s','t','AIC','BIC','HQIC','LLF','MAE','MSE','SSE'])

print("Ordenamos por AIC")
print(orden_df.sort_values('AIC').head())

#print("Ordenamos por BIC")
#print(orden_df.sort_values('BIC').head())
```
